```python
df = pd.read_csv("archive (1).zip")
```

```python
# Display basic info
print("Dataset Info:\n")
df.info()
print("\nFirst 5 rows:\n", df.head())
```

```python
# Convert necessary columns to numeric
numeric_columns = ['total_runs', 'total_balls_faced', 'total_wickets_taken', 'total_matches_played', 'matches_won', 'matches_lost']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
# Calculate the average total runs scored by players
average_runs = df['total_runs'].mean()
print(f"\nAverage Total Runs Scored: {average_runs:.2f}")
```

```python
# Visualization 1: Top 10 Run Scorers
top_scorers = df[['player_name', 'total_runs']].sort_values(by='total_runs', ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='total_runs', y='player_name', data=top_scorers, palette='viridis')
plt.xlabel("Total Runs")
plt.ylabel("Player Name")
plt.title("Top 10 Run Scorers in ODI Cricket")
plt.show()
```

```python
# Visualization 2: Scatter Plot (Total Runs vs. Matches Played)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['total_matches_played'], y=df['total_runs'], alpha=0.6)
plt.xlabel("Total Matches Played")
plt.ylabel("Total Runs Scored")
plt.title("Total Runs vs. Matches Played")
plt.show()
```

```python
# Visualization 3: Heatmap of Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Player Statistics")
plt.show()
```
